```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import the data
our_data = pd.read_csv("insurance.csv")
our_data.head()
```

```python
# The iloc function allows use to select rows from the dataframe for plotting.  
x = our_data.iloc[:,:1].values
y = our_data.iloc[:,-1].values

# Plot a scatter plot with age on the x-axis and charges on the y-axis.
plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Relationship between Age and Charges')
plt.show()
```

```python
# Fit the model
insurance_cost = LinearRegression()
insurance_cost.fit(x,y)
```

```python
# Plot the data and model
y_pred = insurance_cost.predict(x)

# Plot a scatter plot with age on the x-axis and charges on the y-axis with Best-fit line.
plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Relationship between Age and Charges with Best-fit Line')
plt.show()
```

```python
# Predict an unknown value
unk_x = [[70]] 

x_pred = np.append(x, unk_x).reshape(-1,1)
y_pred = insurance_cost.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Relationship between Age and Charges with Best-fit Line')
plt.show()

print("Insurance cost for the selected age is:", insurance_cost.predict(unk_x))
```
